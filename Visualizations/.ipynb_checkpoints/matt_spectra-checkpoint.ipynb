## Functions I might use

```python
def specint(wn, spec, filtwn_list, filttrans_list): #from Ryan Challener's Theresa code
    """
    Integrate a spectrum over the given filters.
    Arguments
    ---------
    wn: 1D array
        Wavenumbers (/cm) of the spectrum 
        Hayley's note: be wary of these units! 
    spec: 1D array
        Spectrum to be integrated
    filtwn_list: list
        List of arrays of filter wavenumbers, in /cm.
    filttrans_list: list
        List of arrays of filter transmission. Same length as filtwn_list.
        Hayley's note: for now, this can just be an array you create of 1's and 0's. (1 where want to integrate over, 0 outside)
    Returns
    -------
    intspec: 1D array
        The spectrum integrated over each filter. 
    """
    if len(filtwn_list) != len(filttrans_list):
        print("ERROR: list sizes do not match.")
        raise Exception
    
    intspec = np.zeros(len(filtwn_list)) 
    
    for i, (filtwn, filttrans) in enumerate(zip(filtwn_list, filttrans_list)):
        # Sort ascending
        idx = np.argsort(filtwn)
        
        intfunc = spi.interp1d(filtwn[idx], filttrans[idx],
                               bounds_error=False, fill_value=0)

        # Interpolate transmission
        inttrans = intfunc(wn)

        # Normalize to one
        norminttrans = inttrans / np.trapz(inttrans, wn)
        

        # Integrate filtered spectrum
        intspec[i] = np.trapz(spec * norminttrans, wn)

    return intspec


def getSpacing(arr):
    return (arr[-1]-arr[0])/(len(arr)-1)

def fwhm2sigma(fwhm):
    return fwhm / np.sqrt(8 * np.log(2))

def gaussian(x, mu, sig):
    normalization = 1/np.sqrt(2 * np.pi * sig**2)
    exponent = - ((x - mu)**2 / (2*sig**2))
    y = normalization * np.exp(exponent)
    return y

def reduceSpectralResolution(x, y, R_low, R_high=None, lambda_mid=None, n=4):
    if np.ndim(y) == 2:
        return np.array([reduceSpectralResolution(x,spec,R_low,R_high,lambda_mid,n) for spec in y])

    dx = getSpacing(x)
    # If lambda_mid is none, take median of input wavelengths
    if lambda_mid is None:
        lambda_mid = np.median(x)

    # If R_high is none, use midpoint of x divided by spacing in x
    if R_high is None:
        R_high = lambda_mid/dx

    # Create Gaussian kernel
    fwhm = np.sqrt(R_high**2 - R_low**2)*(lambda_mid/(R_low*R_high))
    sigma = fwhm2sigma(fwhm)

    kernel_x = np.arange(-n*sigma, n*sigma+dx, dx)
    kernel = gaussian(kernel_x,0,sigma)
    kernel = kernel/np.sum(kernel)

    # find center of kernel
    n_kernel_lt0 = len(np.where(kernel_x<0)[0])
    n_kernel_gt0 = len(np.where(kernel_x>0)[0])

    if n_kernel_lt0< n_kernel_gt0:
        origin = 0
    else:
        origin=-1

    # convolve
    lowRes = ndimage.convolve(y, kernel, origin=origin)
    return lowRes
```

## Converting emission curves to relative flux units

To convert these phase curves into relative flux units, we need the integrated flux from a stellar model and the transit depth.

```python
# Loading in model spectrum of WASP-43
# computed from http://svo2.cab.inta-csic.es/theory/newov2/index.php
starmodel = np.loadtxt('/home/matthew/data/wasp43b/stellar_spectrum/lte043.0-4.5-0.0a+0.0.BT-Settl.spec.7.dat.txt',
                      skiprows=8, unpack=True)
star_wavs = starmodel[0]   # [angstrom]
star_flux = starmodel[1]   # [erg/s/cm2/A]
del starmodel

# Unpacking only the 1 - 10 micron portion
idxs = np.where((star_wavs >= 10000) & (star_wavs <= 100000))
star_wavs = star_wavs[idxs]
star_flux = star_flux[idxs]
```

```python
# Loading in the Spitzer IRAC ch2 response function
response = np.loadtxt('/home/matthew/data/spitzer/spitzerirac_ch2_responsefunction.txt', unpack=True)
irac_respwavs = response[0]*10000   # wavelength in [A]
irac_respfunc = response[1]         # value of response function 
```

```python
# the star_wavs array is much longer, and contains more finely spaced wavelengths, than
#     does the irac_respwavs array
# but ... we need a star_wavs array that is matched up to the irac_respwavs in order to
#    apply the filter response function
# here, I try matching them by going through each filter function wavelength, finding
#    the point in the stellar model arrays that are closest to that particular wavelength
#    and save that wavelength and stellar flux
IRAC_wavs = np.zeros(len(irac_respwavs))      # to hold matched wavelengths
IRAC_modelflux = np.zeros(len(irac_respwavs)) # to hold stellar model's flux at these wavelengths

for i, iracwave in enumerate(irac_respwavs):
    # computing the closest wavelength in star model array to current response func wave
    wave_diffs = abs(star_wavs - iracwave)
    closest_idx = np.argmin(wave_diffs)
    # appending to our ch2-bandpass specific list
    IRAC_wavs[i] = star_wavs[closest_idx]
    IRAC_modelflux[i] = star_flux[closest_idx]
# Now applying the response function to the fluxes
IRAC_modelflux *= irac_respfunc
```

```python
from scipy.integrate import trapz
```

```python
Wm2_to_ergscm2 = 10000000. / 10000.  # conversion factor
# Computing the integrated stellar surface flux over whole IRAC ch2 bandpass
#   compute in [erg/s/cm2]
# using the stellar model (response function already multiplied in!!):
surfaceflux_irac = trapz(IRAC_modelflux, x=IRAC_wavs)
# convert to W/m2
surfaceflux_irac_Wm2 = surfaceflux_irac / Wm2_to_ergscm2
print('band integrated stellar surface flux = %.1f erg/s/cm2 =  %.1f W/m^2'%(surfaceflux_irac, surfaceflux_irac_Wm2))
```

```python
## Get the transit depth from Rp/Rs
## just using the best-fit value from my observations
RpRs = 0.1579
```

Relative flux = $\frac{\text{band-integrated planet flux in W/m2}}{\text{band-integrated stellar flux in W/m2}} * (\text{transit depth})$

```python
## Going through and converting each [W/m^2] array to relative
##    flux units
## Doing just the 4.5 micron curves here 

# arrays to hold relative flux 
relfluxes_all_ch2 = np.zeros(intfluxes_a_Wm2_45.shape)
relfluxes_nl_ch2 = np.zeros(intfluxes_nl_Wm2_45.shape)

# all clouds:
for i, farray in enumerate(intfluxes_a_Wm2_45):
    # compute relative flux
    relflux_a = (farray / surfaceflux_irac_Wm2)*(RpRs**2)
    # store to array
    relfluxes_all_ch2[i] = relflux_a*1.e6
    
# nucleation limited:
for i, farray in enumerate(intfluxes_nl_Wm2_45):
    # compute relative flux
    relflux_nl = (farray / surfaceflux_irac_Wm2)*(RpRs**2)
    relfluxes_nl_ch2[i] = relflux_nl*1.e6
    
```

```python
## Now plotting them to see how it looks
fig, ax = plt.subplots(figsize=(10,8), nrows=2, sharex=True)
plt.subplots_adjust(hspace=0.1)
ax0, ax1 = ax[0], ax[1]

labels = ['5 Cloud Layers',
          '10 Cloud Layers',
          '15 Cloud Layers',
          '20 Cloud Layers',
          '25 Cloud Layers',
          '30 Cloud Layers',
          '35 Cloud Layers',
          '40 Cloud Layers',
          '45 Cloud Layers']

# all clouds:
for i, farray in enumerate(intfluxes_a_Wm2_45):
    # plot the planet's emission curve
    ax0.plot(phases_a, farray, label=labels[i])
    # plot the relative flux curve
    ax1.plot(phases_a, relfluxes_all_ch2[i], label=labels[i])
# nucleation limited:
for i, farray in enumerate(intfluxes_nl_Wm2_45):
    # plot the planet's emission curve
    ax0.plot(phases_nl, farray, ls='--')
    # plot the relative flux curve
    ax1.plot(phases_nl, relfluxes_nl_ch2[i], ls='--')
    
    
ax0.set(xlabel='phase', ylabel='flux [W/m^2]')
ax0.legend(loc='upper right', fontsize=8)
ax1.set_ylim(0, 6000)
ax1.set_xlim(0,1)
ax1.set(xlabel='phase', ylabel='relative flux [ppm]')
ax1.legend(loc='upper right', fontsize=8)
print('Note: solid lines = all clouds; dashed lines = nucl. limited species only')
plt.show()
```
