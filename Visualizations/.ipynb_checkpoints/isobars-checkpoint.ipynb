```python
cases = [file1]


column_names = ['lat', 'lon', 'level',
               'alt', 'pres', 'temp', 
               'u', 'v', 'w',
               'aero_sw_tau_1', 'sw_asym_1', 'sw_pi0_1',
               'aero_sw_tau_2', 'sw_asym_2', 'sw_pi0_2',
               'aero_sw_tau_3', 'sw_asym_3', 'sw_pi0_3',
               'aero_sw_tau_4', 'sw_asym_4', 'sw_pi0_4',
               'aero_sw_tau_5', 'sw_asym_5', 'sw_pi0_5',
               'aero_sw_tau_6', 'sw_asym_6', 'sw_pi0_6',
               'aero_sw_tau_7', 'sw_asym_7', 'sw_pi0_7',
               'aero_sw_tau_8', 'sw_asym_8', 'sw_pi0_8',
               'aero_sw_tau_9', 'sw_asym_9', 'sw_pi0_9',
               'aero_sw_tau_10', 'sw_asym_10', 'sw_pi0_10',
               'aero_sw_tau_11', 'sw_asym_11', 'sw_pi0_11',
               'aero_sw_tau_12', 'sw_asym_12', 'sw_pi0_12',
               'aero_sw_tau_13', 'sw_asym_13', 'sw_pi0_13']
nlat = 48
nlon = 193
nlevel = 60
nparams = 48    
    
for case in cases:
    df = pd.read_csv(case + '.txt', delim_whitespace=True, names=column_names)

    # Double the data
    double = df.copy()
    double.lon = double.lon + 360.0

    double2 = df.copy()
    double2 = double2[(double2['lon'] == 0.0)]
    double2.lon = double2.lon + 720.0

    doubled = pd.concat([df, double, double2])
    doubled = doubled.sort_values(by=['lat', 'lon', 'level'], axis=0, ascending=[True, True, True])

    numpy_df = doubled.to_numpy()
    np.savetxt(case + '_doubled.txt', numpy_df,
            fmt=' '.join(['%5.2f']*2 + ['%3d']*1 + ['%9.2E']*6 + ['%9.2E']*39 + ['\t']))
```

```python
### ----- INPUT/OUTPUT CONTROL ----- ###

fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(12,9))
plt.subplots_adjust(wspace=0.01, hspace=0.015)

P_phots = [1e-2] # Photospheric pressure in bars

for ind, case in enumerate(cases):
    P_phot = P_phots[ind]
    
    nlat = 48
    nlon = 193
    nlevel = 60
    nparams = 48
        
    data = np.loadtxt(case + '_doubled.txt')
    data = data.reshape((nlat, nlon, nlevel, nparams))
    
    # need to get indices for given pressure level
    pressure_ind = np.zeros((nlat, nlon))

    for i in range(nlat):
        for j in range(nlon):
            k = nlevel - 1

            while data[i][j][k][4] > P_phot:
                k -= 1
            pressure_ind[i][j] = k
            
        lons = data[:,:,0,1][0] - 360
    lats = data[:,:,0,0][:,0]
    

    ### ----- GET LOCAL TEMP, AEROSOL TAU, etc... ----- ###

    print('Calculating T and aerosol tau...')

    temps = np.zeros((nlat, nlon))
    taus1 = np.zeros((nlat, nlon))
    taus2 = np.zeros((nlat, nlon))
    taus3 = np.zeros((nlat, nlon))
    taus4 = np.zeros((nlat, nlon))
    taus5 = np.zeros((nlat, nlon))
    taus6 = np.zeros((nlat, nlon))
    taus7 = np.zeros((nlat, nlon))
    taus8 = np.zeros((nlat, nlon))
    taus9 = np.zeros((nlat, nlon))
    taus10 = np.zeros((nlat, nlon))
    taus11 = np.zeros((nlat, nlon))
    taus12 = np.zeros((nlat, nlon))
    taus13 = np.zeros((nlat, nlon))
    EW_vels = np.zeros((nlat, nlon))
    NS_vels = np.zeros((nlat, nlon))
    vert_vels = np.zeros((nlat, nlon))
    pressures = np.zeros((nlat, nlon))
    z = np.zeros((nlat, nlon))

    ### ----- CALCULATE TEMP INVERSIONS FOR PLOTTING ----- ###

    # empty array to store values
    delta_temps = np.zeros((nlat, nlon))
    taus = np.zeros((nlat, nlon))

    for i in range(nlat):
        for j in range(nlon):

            z[i][j] = data[i][j][int(pressure_ind[i][j])][3]
            pressures[i][j] = data[i][j][int(pressure_ind[i][j])][4]
            temps[i][j] = data[i][j][int(pressure_ind[i][j])][5]
            EW_vels[i][j] = data[i][j][int(pressure_ind[i][j])][6]
            NS_vels[i][j] = data[i][j][int(pressure_ind[i][j])][7]
            vert_vels[i][j] = data[i][j][int(pressure_ind[i][j])][8]


            # integrate aerosol optical depth above pressure level
            k = 0

            while k <= int(pressure_ind[i][j]):

                #taus1[i][j] += data[i][j][k][9]
                #taus2[i][j] += data[i][j][k][12]
                #taus3[i][j] += data[i][j][k][15]
                #taus4[i][j] += data[i][j][k][18]
                
                #taus5[i][j] += data[i][j][k][21]
                #taus6[i][j] += data[i][j][k][24]
                #taus7[i][j] += data[i][j][k][27]
                #taus8[i][j] += data[i][j][k][30]
                
                #taus9[i][j] += data[i][j][k][33]
                #taus10[i][j] += data[i][j][k][36]
                #taus11[i][j] += data[i][j][k][39]
                #taus12[i][j] += data[i][j][k][42]
                #taus13[i][j] += data[i][j][k][45]
                
                taus1[i][j] += data[i][j][k][9]
                taus1[i][j] += data[i][j][k][12]
                taus1[i][j] += data[i][j][k][15]
                taus1[i][j] += data[i][j][k][18]
                
                taus1[i][j] += data[i][j][k][21]
                taus1[i][j] += data[i][j][k][24]
                taus1[i][j] += data[i][j][k][27]
                taus1[i][j] += data[i][j][k][30]
                
                taus1[i][j] += data[i][j][k][33]
                taus1[i][j] += data[i][j][k][36]
                taus1[i][j] += data[i][j][k][39]
                taus1[i][j] += data[i][j][k][42]
                taus1[i][j] += data[i][j][k][45]
            

                # The clouds have extremely high optical depths
                k += 1
                

    # stream plot
    test_x = np.linspace(min(lons), max(lons), len(lons))
    test_y = np.linspace(min(lats), max(lats), len(lats))

    temp_map = axes.contourf(lons, lats, temps, cmap=temperature_colors, levels=100)      
    flow = axes.streamplot(test_x, test_y, EW_vels, NS_vels, linewidth=1.2,density=([1,1.5]),color='#d8dcd6', zorder=1)
    
    #cloud_map = axes[1].contourf(lons, lats, taus1, cmap=cloud_colors)
    
    
    #cloud_map = axes[2].contourf(lons, lats, taus2, cmap=cloud_colors)
    #cloud_map = axes[3].contourf(lons, lats, taus3, cmap=cloud_colors)
    #cloud_map = axes[4].contourf(lons, lats, taus4, cmap=cloud_colors)
    #cloud_map = axes[5].contourf(lons, lats, taus5, cmap=cloud_colors)
    #cloud_map = axes[6].contourf(lons, lats, taus6, cmap=cloud_colors)
    #cloud_map = axes[7].contourf(lons, lats, taus7, cmap=cloud_colors)
    #cloud_map = axes[8].contourf(lons, lats, taus8, cmap=cloud_colors)
    #cloud_map = axes[9].contourf(lons, lats, taus9, cmap=cloud_colors)
    #cloud_map = axes[10].contourf(lons, lats, taus10, cmap=cloud_colors)
    #cloud_map = axes[11].contourf(lons, lats, taus11, cmap=cloud_colors)
    #cloud_map = axes[12].contourf(lons, lats, taus12, cmap=cloud_colors)
    #cloud_map = axes[13].contourf(lons, lats, taus13, cmap=cloud_colors)
    
    


# format axes
#for ax in axes:#.flatten():
axes.set_xlim([-180, 180])
axes.set_ylim([-87, 87])
axes.spines['left'].set_position('zero')
axes.spines['bottom'].set_position('zero')
axes.spines['left'].set_color('none')
axes.spines['right'].set_color('none')
axes.spines['top'].set_color('none')
axes.spines['bottom'].set_color('none')
axes.tick_params(
    axis='both',
    which='both',
    top=False,
    bottom=False,
    right=False,
    left=False,
    labelcolor='w',
    labelsize=25,
    pad=-12)
axes.yaxis.get_major_ticks()[2].label1.set_visible(False)
axes.grid(color='w', alpha=0.5, ls=':')


axes.patch.set_edgecolor('black')  
axes.patch.set_linewidth('2')  
axes.grid(color='w', alpha=0.5, ls=':')

temp_cbar = fig.colorbar(temp_map, location='top', aspect=30, pad=0.005) #ax=axes.ravel().tolist(),
#temp_cbar = fig.colorbar(temp_map, aspect=30, pad=0.005)

temp_cbar.set_label('Temperature (K)', fontsize=26)

cloud_cbar = fig.colorbar(cloud_map, ax=axes.ravel().tolist(), location='bottom',aspect=30, pad=0.005)
cloud_cbar.set_label('Cloud Optical Depth, 2.3 $\\mu$m', fontsize=26)

plt.savefig('../Figures/GJ1214b/Cloud_Coverage_Isobars_{}_bar_{}.png'.format(P_phots[0], planet_name), bbox_inches='tight', dpi=250)
```
