```python
### ----- IMPORT LIBRARIES ----- ###

import numpy as np 
import matplotlib as mpl
from matplotlib import rcParams, rc
import pandas as pd
import matplotlib.pyplot as plt 
import matplotlib.colors as mcolors
from scipy.interpolate import interp2d
import matplotlib.colors as colors


plt.style.use('../../isaac-style.txt')
```

```python
### ----- INPUT/OUTPUT CONTROL ----- ###

cases = ['../Spectra-Processing/Spectra/DATA/init_HD209-Table1-Ya-Clouds-Cor0_phase_0.0_inc_0.0',
         '../Spectra-Processing/Spectra/DATA/init_HD209-Table1-Ya-Clouds-Cor0-Compact_phase_0.0_inc_0.0',
         '../Spectra-Processing/Spectra/DATA/init_HD209-Table1-Ya-Clouds-Cor1_phase_0.0_inc_0.0',
         '../Spectra-Processing/Spectra/DATA/init_HD209-Table1-Ya-Clouds-Cor1-Freedman_phase_0.0_inc_0.0']


fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(20,10), sharex=True, sharey=True)
plt.subplots_adjust(wspace=0.02, hspace=0.03)


# file info

temp_low = 1000
temp_high = 2250

clouds_low = 0.0
clouds_high = 1.0


P_phots = [80.0, 80.0, 80.0, 80.0]   


for ind, case in enumerate(cases):
    P_phot = P_phots[ind]
    
    
    
    if True:
        nlat = 48
        nlon = 193
        nlevel = 100
        nparams = 22

        # This is terrible code
        # I'm so sorry
        column_names = ['lat' , 'lon', 'level' ,
                        'altitude(m)','pressure(bars)', 'temp(k)',
                        'EW vel(m/s)','NS vel', 'w',
                        'aero_sw_tau_1', 'sw_asym_1', 'sw_pi0_1',
                        'aero_sw_tau_2', 'sw_asym_2', 'sw_pi0_2',
                        'aero_sw_tau_3', 'sw_asym_3', 'sw_pi0_3',
                        'aero_sw_tau_4', 'sw_asym_4', 'sw_pi0_4',
                        'incident_frac']

        df = pd.read_csv(case + '.txt', delim_whitespace=True, names=column_names)

        # Double the data
        double = df.copy()
        double.lon = double.lon + 360.0

        double2 = df.copy()
        double2 = double2[(double2['lon'] == 0.0)]
        double2.lon = double2.lon + 720.0

        doubled = pd.concat([df, double, double2])
        doubled = doubled.sort_values(by=['lat', 'lon', 'level'], axis=0, ascending=[True, True, True])

        numpy_df = doubled.to_numpy()
        np.savetxt(case + '_Doubled.txt', numpy_df,
                    fmt='%12.4E  %12.4E  %12.4E  %12.4E  %12.4E  %12.4E  %12.4E  %12.4E  %12.4E  %12.4E  %12.4E  %12.4E  %12.4E  %12.4E  %12.4E  %12.4E  %12.4E  %12.4E  %12.4E  %12.4E  %12.4E %12.4E\t')

        data = np.loadtxt(case + '_Doubled.txt')
        data = data.reshape((nlat, nlon, nlevel, nparams))
    else:
        nlat = 48
        nlon = 96
        nlevel = 100
        nparams = 22
        
        data = np.loadtxt(case + '.txt')
        data = data.reshape((nlat, nlon, nlevel, nparams))

    # need to get indices for given pressure level
    pressure_ind = np.zeros((nlat, nlon))

    for i in range(nlat):
        for j in range(nlon):

            k = nlevel - 1

            while data[i][j][k][4] * 1e-2 > P_phot:
                k -= 1

            pressure_ind[i][j] = k


    lons = data[:,:,0,1][0] - 180.
    lats = data[:,:,0,0][:,0]

    ### ----- GET LOCAL TEMP, AEROSOL TAU, etc... ----- ###

    print('Calculating T and aerosol tau...')

    temps = np.zeros((nlat, nlon))
    taus = np.zeros((nlat, nlon))
    EW_vels = np.zeros((nlat, nlon))
    NS_vels = np.zeros((nlat, nlon))
    vert_vels = np.zeros((nlat, nlon))
    pressures = np.zeros((nlat, nlon))
    z = np.zeros((nlat, nlon))
    

    ### ----- CALCULATE TEMP INVERSIONS FOR PLOTTING ----- ###

    # empty array to store values
    delta_temps = np.zeros((nlat, nlon))
    taus = np.zeros((nlat, nlon))

    for i in range(nlat):
        for j in range(nlon):

            z[i][j] = data[i][j][int(pressure_ind[i][j])][3]
            pressures[i][j] = data[i][j][int(pressure_ind[i][j])][4]
            temps[i][j] = data[i][j][int(pressure_ind[i][j])][5]
            EW_vels[i][j] = data[i][j][int(pressure_ind[i][j])][6]
            NS_vels[i][j] = data[i][j][int(pressure_ind[i][j])][7]
            vert_vels[i][j] = data[i][j][int(pressure_ind[i][j])][8]


            # integrate aerosol optical depth above pressure level
            k = 0

            # Something really funky is going on here
            while k <= int(pressure_ind[i][j]):

                # conversion to 2.3 microns
                taus[i][j] += data[i][j][k][9] * (0.07 / 0.01)  * 0.2
                taus[i][j] += data[i][j][k][12] * (1.25 / 0.16) * 0.2
                taus[i][j] += data[i][j][k][15] * (0.12 / 0.02) * 0.2
                taus[i][j] += data[i][j][k][18] * (0.56 / 0.02) * 0.2

                # The clouds have extremely high optical depths
                k += 1
    

    print('Calculating T and aerosol tau DONE\n')

    # temp colormap
    cm_name = 'lajolla'
    cm_file = np.loadtxt(f'../../ScientificColourMaps7/{cm_name}/{cm_name}.txt')
    cm_file = np.flip(cm_file, axis=0)
    my_colors = mcolors.LinearSegmentedColormap.from_list(cm_name, cm_file)

    print ('temp:', np.amin(temps), np.amax(temps))
    temp_map = axes[0][ind].contourf(lons-180.0, lats, temps,
                                     cmap=my_colors, levels=np.arange(temp_low, temp_high, 10))
    temp_map = axes[0][ind].contourf(lons-180.0, lats, temps,
                                     cmap=my_colors, levels=np.arange(temp_low, temp_high, 10))

    # stream plot
    test_x = np.linspace(min(lons), max(lons), len(lons))
    test_y = np.linspace(min(lats), max(lats), len(lats))

    flow = axes[0][ind].streamplot(test_x, test_y, EW_vels, NS_vels, linewidth=1.,
                                   density=([1, 3]),color='#d8dcd6', zorder=1)

    # cloud colormap
    cm_name = 'devon' 
    cm_file = np.loadtxt(f'../../ScientificColourMaps7/{cm_name}/{cm_name}.txt')
    my_colors = mcolors.LinearSegmentedColormap.from_list(cm_name, cm_file)


    # interpolate clouds for nicer plotting
    interp_cloud = interp2d(lons, lats, taus, kind='linear')
    lons_new = np.linspace(-180,180,30)
    lats_new = np.linspace(-90.,90.,15)
    XX, YY = np.meshgrid(lons_new, lats_new)
    clouds_new = interp_cloud(lons_new, lats_new)

    cloud_map = axes[1][ind].contourf(XX, YY, clouds_new, cmap=my_colors)

    axes[0][ind].set_xticks([-90, 90])
    axes[1][ind].set_xticks([-90, 90])

    axes[0][ind].set_yticks([-50, 50])
    axes[1][ind].set_yticks([-50, 50])

# format axes
i = 0
for ax in axes.flatten():    

    ax.set_xlim([-180, 180])
    ax.set_ylim([-87, 87])
    ax.spines['left'].set_position('zero')
    ax.spines['bottom'].set_position('zero')
    ax.spines['left'].set_color('none')
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.spines['bottom'].set_color('none')


    

    if i < 4:
        ax.tick_params(
            axis='both',
            which='both',
            top=False,
            bottom=False,
            right=False,
            left=False,
            labelcolor='black',
            labelleft=True,
            labelbottom=True,
            size=10,
            pad=-12)
    else:
        ax.tick_params(
            axis='both',
            which='both',
            top=False,
            bottom=False,
            right=False,
            left=False,
            labelcolor='white',
            labelleft=True,
            labelbottom=True,
            size=10,
            pad=-12)
        
    ax.patch.set_edgecolor('black')  
    ax.patch.set_linewidth('2')  

    
    ax.grid(color='w', alpha=0.5, ls=':')
    
    i = i + 1



temp_cbar = fig.colorbar(temp_map, ax=axes.ravel().tolist(), 
                         norm=colors.LogNorm(vmin=0, vmax=1000),
                         location='top', aspect=80, pad=0.02)
temp_cbar.set_label('Temperature (K)', weight='bold', fontsize=26)

cloud_cbar = fig.colorbar(cloud_map, ax=axes.ravel().tolist(), location='bottom',
                          aspect=80, pad=0.02)
cloud_cbar.set_label('Cloud Optical Depth, 2.3 $\\mu$m', weight='bold', fontsize=26)

print('Creating plot DONE')


plt.savefig('../Figures/surface-winds.png', bbox_inches='tight', dpi=250)
```
